```python
import pandas as pd
import numpy as np
marketing_customer_analysis = pd.read_csv('./files_for_lab/csv_files/marketing_customer_analysis.csv')
marketing_customer_analysis
```

```python
# Check the data types of the columns
marketing_customer_analysis.dtypes
```

```python
# Get the numeric data into dataframe called numerical and categorical columns in a dataframe called categoricals.
numerical = marketing_customer_analysis.select_dtypes(include = np.number)
categoricals = marketing_customer_analysis.select_dtypes(include = np.object)
display(numerical)
categoricals
```

```python
# Use seaborn library to construct distribution plots for the numerical variables
import seaborn as sns
import matplotlib.pyplot as plt
for column in numerical.columns:
    sns.displot(numerical[column])
    plt.title(f"Distribution of {column}")
    plt.show()
```

```python
# Use Matplotlib to construct histograms
for column in numerical.columns:
    plt.hist(numerical[column], bins=50)
    plt.title(f"Distribution of {column}")
    plt.show()
```

```python
# Do the distributions for different numerical columns look symmetrical? Compute the skewness for each, and add a comment with 
# your findings.
for column in numerical.columns:
    print(numerical[column].skew())
print("Most of the distributions for the different numerical columns don't look symmetrical. All of them are positively skewed to different extent. Some of the distributions are highly skewed and deviate significantly from a symmetric shape because their skewness result is higher than 2. The columns that belong to this case are Customer Lifetime Value, Monthly Premium Auto and Number of Open Complaints. The rest of the column distributions have a skewness inside the [-2,2] range, so their distributions are approximately symmetric or only have a minor degree of skewness. The columns with the most symmetrical distributions are Income, Months Since Last Claim and Months Since Policy Inception, especially the latter.")
```

```python
# For the numerical variables, check the multicollinearity between the features.
correlation_matrix = numerical.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
print("High correlations indicate potential multicollinearity, so by looking at this heatmap of the correlation matrix we can conclude that no columns have multicollinearity with another column because no correlation is high enough (greater than 0.9).")
```
